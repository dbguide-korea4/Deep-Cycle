```python
import requests
import os
```

```python
from selenium import webdriver
driver = webdriver.Chrome()
```

```python
driver.get("https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%ED%8E%98%ED%8A%B8%EB%B3%91")
```

```python
n=0
while n!=1000:
    n=n+1
    driver.find_elements_by_css_selector("div.photo_grid._box div.img_area._item")[n].click()
    for _ in driver.find_elements_by_css_selector("div.viewer img"):
        temp1=_.get_attribute('src').split('&')[0]
        temp2=temp1.split('.')[-1]
        image=requests.get(temp1).content    
        with open(os.path.join(f'./img/img{n}.{temp2}'),'wb') as fp:
            fp.write(image)
            print('sucess')
       
    
    
    
```
